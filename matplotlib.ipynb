```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
###Draw a line in a diagram from position(0,10) to position(5,150):
xpoints=np.array([1,2])
ypoints=np.array([5,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,20])
ypoints=np.array([10,200])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,6])
ypoints=np.array([4,5])
plt.plot(xpoints,ypoints,"X")
plt.show 
```

```python
ypoints=np.array([83,8,1,10])
plt.plot(ypoints,"*""r")
plt.show()
```

```python
ypoints=np.array([3,8,1,12])
xpoints=np.array([5,6,7,9])
plt.plot(xpoints,ypoints,marker="*",ms=30,color="red")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="o",ms=30,mec="r",color="black")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="o",ms=20,mfc="g",mec="k")
plt.show()
```

```python
###use both the mec and mfc arguments to color of the entire marker
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linestyle="dashed",marker="o",mec="hotpink",ms=20,mfc="blue",linewidth=3)
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="o",ms=200,mec="y",mfc="k")
plt.show()
```

```python
plt.plot(ypoints,linestyle="dotted",marker="o",ms=30,mfc="green",mec="red")
```

```python
xpoints=np.array([10,20,30,40])
ypoints=np.array([1,2,6,8])
plt.plot(xpoints,ypoints,marker="*",color="yellow",mfc="g",mec="r",ms=20,linestyle="dashed")
```
